```python
suicides_vs_gpd = df[['suicides_no', 'year', 'gdp_per_capita ($)']].groupby('year').mean()
suicides_vs_gpd.reset_index(inplace=True)

suicides_vs_gpd
```

```python
fig, ax1 = plt.subplots()

# Plot the suicides over the years.
lns1 = ax1.plot(suicides_vs_gpd['year'], suicides_vs_gpd['suicides_no'], 'C0', label='suicides_no')

# Create a shared axis for plotting on a different scale the GPD.
ax2 = ax1.twinx()
lns2 = ax2.plot(suicides_vs_gpd['year'], suicides_vs_gpd['gdp_per_capita ($)'], 'C1', label='GPD')

# Join both legends into the same box.
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=2)

# Set the labels.
ax1.set_ylabel('Suicides per 100k population')
ax2.set_ylabel('GDP per Capita')
ax1.set_xlabel('Years')

plt.tight_layout()
plt.show()
```
